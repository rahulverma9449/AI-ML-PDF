```python
# How to find pairs in an array whose sum is equal to the given number?
# Example - if the given array is (2,6,3,9,11) and the given number is 9, output would be (6,3)

def findpair(arr, sum):
    pair = []
    n  = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] + arr[j] == sum:
                pair.append((arr[i], arr[j]))
    return pair


arr = [2,6,3,9,11]
sum = 9
print(findpair(arr, sum))
```

```python
def removeduplicate(arr):
    if not arr:
        return 0  # Edge case: empty array

    size = len(arr)
    index = 1  # Start from the second element
    for i in range(1, size):
        if arr[i-1] != arr[i]:  # Check if current element is not the same as the previous one
            arr[index] = arr[i]  # Move the unique element to the 'index' position
            index += 1  # Increment the index for the next unique element

    return index  # Return the count of unique elements

arr = [1, 2, 2, 3, 3]
result = removeduplicate(arr)
print("Count of unique elements:", result)
print("Modified array with unique elements:", arr[:result])
```

```python
#Given an integer array nums and an integer k, return the k most frequent elements. You may return the answer in any order.

# Given array of integers and the value of k
nums = [1, 1, 1, 2, 2, 3]
k = 2

# Count the frequency of each element
count = Counter(nums)

# Get the k most common elements
common = count.most_common(k)

# Extract the elements from the tuples returned by most_common()
result = [item for item, freq in common]

# Output the result
print(result)  # Output: [1, 2]

```

```python
from collections import Counter
def top_freq(nums, k):
  count = Counter(nums)
  common = count.most_common(k)
  return [item for item, freq in common]

nums = [1,1,1,2,2,3]
result = top_freq(nums, 2)
print(result)
```

```python
def top_k_freq(nums, k):
  freq = {}
  for num in nums:
    freq[num] = freq.get(num, 0)+1
  sorted_item = sorted(freq.items(), key=lambda item: item[1], reverse=True)
  return [item[0] for item in sorted_item[:k]]

nums = [1,1,1,2,2,3]
k = 2
result = top_k_freq(nums, k)
print(result)

```

```python
# Write the code without function?

arr = [2,6,3,9,1]
pair = []
target = 9
seen = set()

for num in arr:
    complement = target - num
    if complement in seen:
        pair.append((complement, num))
    seen.add(num)

print(pair)
```

```python
# Decorator is a function that takes another function as agrument and returns a function.
# """In Python, decorators are a powerful feature that allow you to modify the behavior of a function or method.
# There are several built-in decorators, such as @staticmethod, @classmethod, and @property.
# These decorators are often used to define methods in classes that have special behaviors."""

def decorator(func):
  def wrapper():
    print("The real estate project is completed")
    func()
    print("Team are working on this ")
  return wrapper

@decorator
def hello():
  print("Hello world")

hello()


```

```python
from time import sleep
from threading import Thread

# Class A inherits from threading.Thread
class A(Thread):
    def run(self):
        for i in range(5):
            print("executed")  # "executed" should be a string
            sleep(1)

# Class B inherits from threading.Thread
class B(Thread):
    def run(self):
        for i in range(5):
            print("Thread B running")  # Placeholder print statement
            sleep(1)  # sleep() should take an argument for time in seconds

# Create instances of A and B
t1 = A()
t2 = B()

# Start both threads
t1.start()
t2.start()

# Join the threads to the main thread to ensure they complete
t1.join()
t2.join()
```

```python
# Simple Way of Generator
def generator():
  i = 1
  while i <= 10:
    yield i
    i += 1

# Create a generator object
x = generator()
print(next(x))
print(next(x))
print(next(x))


def generator(n):
  for i in range(n):
    yield i

gen = generator(5)

for num in gen:
  print(num)

```
